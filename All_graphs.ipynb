# **1.Scatter plot : Used to check data linear or non linear**

Scatter plot: Show the relationship between two numerical variables with dots or other symbols.Scatterplot can be used with several semantic groupings which can help to understand well in a graph. They can plot two-dimensional graphics that can be enhanced by mapping up to three additional variables while using the semantics of hue, size, and style parameters. All the parameter control visual semantic which are used to identify the different subsets. Using redundant semantics can be helpful for making graphics more accessible.

**A)scatter plot by x and y column**

```python
#scatter plot by x and y column
sns.scatterplot(x=df2.Close,y=df2.Open)
```

**B)interchanging the parameters of x and y**

```python
#interchanging the parameters of x and y
sns.scatterplot(y=df2.Close,x=df2.Open)
```

**C)by using matplotlib**

```python
plt.scatter(y=df2.Open,x=df2.Close)
```

**D) If space is present in column name**

```python
#If space is present in column name
sns.scatterplot(y=df2["Close"],x=df2["Open"])
```

**E) "hue" parameter**

```python
#with hue
sns.scatterplot(x=df4["Rainfall(mm)"],y=df4["Rented Bike Count"],hue=df4["Seasons"])
```

**F) with categorical variable**

```python
sns.scatterplot(data=df4,x="Seasons",y="Rented Bike Count")
```

**G)with marker attributes**

```python
sns.scatterplot(data=df4,x="Seasons",y="Rented Bike Count",marker = '+')
```

**H) with style**

```python
#with hue
sns.scatterplot(x=df4["Rainfall(mm)"],y=df4["Rented Bike Count"],hue=df4["Seasons"],style=df4["Seasons"])
```

# **2.Relplot**

Relation plot- Similar to scatter plot but we can visualize joint distribution.

```python
#plot1-scatter plot
sns.scatterplot(x=df4["Rainfall(mm)"],y=df4["Rented Bike Count"])

#plot2-relation plot
sns.relplot(x=df4["Rainfall(mm)"],y=df4["Rented Bike Count"])
plt.show()
```

```python
#with "hue" parameter
#plot1-scatter plot
sns.scatterplot(x=df4["Rainfall(mm)"],y=df4["Rented Bike Count"],hue=df4["Seasons"])

#plot2-relation plot
sns.relplot(x=df4["Rainfall(mm)"],y=df4["Rented Bike Count"],hue=df4["Seasons"])
plt.show()
```

**We can conclude that, scatter plot and relation plot are similar kind of graphs.**

# **3.Regression plot**

Regression plot- Similar to a scatter plot but the only addition is that one regression line is added in the plot which can show the relationship between input and output variables.

```python
#plot additional regression best fit line on scatterplot
sns.regplot(x=df2.Close,y=df2.Open,line_kws={"color": "red"})
```

**This is linear line**

```python
#plot1-scatter plot
sns.regplot(x=df4["Hour"],y=df4["Rented Bike Count"],line_kws={"color": "red"})
```

**This shows non linear behaviour.**

# **5.Joint plot**

Joint plot- Represent bivariate distribution using scatter plot and histogram plot.

```python
#without "hue" parameter
sns.jointplot(data=df4,x="Snowfall (cm)",y="Rented Bike Count")
```

# **14.Heatmap**

Heatmap- plot graph between highly activated values and less activated values. 

```python
correalation=df4.corr()
plt.figure(figsize=(15,15))
sns.heatmap(correalation,cbar=True,square=True,fmt=".1f",annot=True,annot_kws={"size":10}, cmap="Blues")
plt.title("Correlation of input variables to output variable")
```

# **16.Box plot**

Box Plot- Generally box plot is used to find outliers in the dataset. A box plot helps to maintain the distribution of quantitative data in such a way that it facilitates the comparisons between variables or across levels of a categorical variable. The main body of the box plot shows the quartiles and the median’s confidence intervals if enabled. The medians have horizontal lines at the median of each box and while whiskers have vertical lines extending to the most extreme, non-outlier data points and caps are the horizontal lines at the ends of the whiskers.

```python
sns.boxplot(data=df3,x="Total intl charge")
```
